    # criando função exponencial

$$ f\left ( x \right ) = a \cdot e^{\frac{-1}{2\cdot \sigma ^{2}}\cdot\left \| x-p \right \| ^{2} } $$

```python
f1 = lambda a,x,s,p: a * np.exp((-1/(2*(s**2)))*((x-p)**2))
```

#### f(x) para conferência

```python
f1(1,6,1,1)
```

#### Criando valores do eixo x

```python
x = np.linspace(-6.0, 8.0, 100)
x[:4]
```

## RBF 1 Polo

```python
a = 1

p1 = -1
p2 = 0
p3 = 1

s1 = 0.3
s2 = 0.5
s3 = 1
s4 = 2

y11 = f1(a,x,s1,p1)
y21 = f1(a,x,s2,p1)
y31 = f1(a,x,s3,p1)
y41 = f1(a,x,s4,p1)


y10 = f1(a,x,s1,p2)
y20 = f1(a,x,s2,p2)
y30 = f1(a,x,s3,p2)
y40 = f1(a,x,s4,p2)


y1m = f1(a,x,s1,p3)
y2m = f1(a,x,s2,p3)
y3m = f1(a,x,s3,p3)
y4m = f1(a,x,s4,p3)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(x, y11)
plt.plot(x, y21)
plt.plot(x, y31)
plt.plot(x, y41)
plt.title("Polo = -1")
plt.legend([s1,s2,s3,s4])
# plt.plot(x, y, 'ro')
plt.show()

plt.plot(x, y10)
plt.plot(x, y20)
plt.plot(x, y30)
plt.plot(x, y40)
plt.title("Polo = 0")
plt.legend([s1,s2,s3,s4])
# plt.plot(x, y, 'ro')
plt.show()

plt.plot(x, y1m)
plt.plot(x, y2m)
plt.plot(x, y3m)
plt.plot(x, y4m)
plt.title("Polo = +1")
plt.legend([s1,s2,s3,s4])
# plt.plot(x, y, 'ro')
plt.show()
```

## Elementos das Funções de Base Radial

$$f(x_{L}) = \sum_{p=1}^{P}a_{p}\cdot R_{p}\left \| (x_{L}) -(C_{p}) \right \|^{2}+b$$

$f()$ = saídas

$x_{L}$ = entradas

$p$ = polos

$a_{p}$ = sinapses

$R_{p}$ = Função

$x_{L}$ = dados

$-$ = distância

$C_{p}$ = coordenadas

$b$ = ajuste



$$\sigma = \frac{(Distancia Maxima entre Polos)}{2*(Numero de Polos)}$$

## 16.0 Exemplo Aula

```python
# Entradas
x = np.array([2,3,4])
x
```

```python
# Saidas
f = np.array([3,6,5])
f
```

```python
from scipy.spatial import distance
dist = lambda x,y: distance.euclidean(x,y)
```

```python
# Entradas
x = np.array([[1,1],[0,1],[0,0],[1,0]])
x
```

```python
# Polos
polos = np.array([[1,1],[0,0]])
```

```python
# Testando funçāo distancia
dist(x[0],polos[0])
```

```python
# Saidas
f = np.array([0,1,0,1])
f
```

```python
# projetar esse calculo com a turma
for i in range(x.shape[0]):
    print (-1*(dist(x[i],polos[0])**2))
```

```python
matriz = np.zeros([x.shape[0],polos.shape[0]])
for i in range(x.shape[0]):
    for j in range(polos.shape[0]):
        matriz[i][j] = np.exp(-1*(dist(x[i],polos[j])**2))
matriz
```

```python
# adicionar uma dimensão a mais
R = np.append(matriz, np.ones((x.shape[0],1)), axis=1)
R
```

```python
# inverter a matriz e multiplicar pelas saídas
coef = np.dot(np.linalg.pinv(R),f)
coef
```

```python
funcao = lambda x: coef[0]*np.exp(-1*(dist(x,polos[0])**2)) + coef[1]*np.exp(-1*(dist(x,polos[1])**2)) + coef[2]
```

```python
xlist = np.linspace(-3.0, 3.0,100)
xlist.shape
```

```python
# plotando grafico
ylist = np.zeros(xlist.shape[0])
for i in range(xlist.shape[0]):    
    ylist[i] = funcao(xlist[i])
ylist.shape
```
